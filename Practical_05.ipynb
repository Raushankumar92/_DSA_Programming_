```python
# 2D array (matrix)

matrix = []

rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

# Input
for i in range(rows):
    row = []
    for j in range(cols):
        element = int(input(f"Enter element [{i}][{j}]: "))
        row.append(element)
    matrix.append(row)

# Display
print("Matrix is:")
for i in range(rows):
    for j in range(cols):
        print(matrix[i][j], end=" ")
    print()
```
